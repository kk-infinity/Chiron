```python
import json
import re
from flask import Flask, request, jsonify, make_response
import requests
from google.cloud import datastore
```

```python
app = Flask(__name__)


@app.route('/webhook/', methods=['POST'])
def handle():
    print('Inside Handle')
    req = request.get_json(silent=True, force=True)
#     print('Request:')
#     print(json.dumps(req, indent=4))
#     if req['queryResult']['action'] != 'assign-rem':
#         return {}
    client = datastore.Client()
    key = client.key('store_val')
    entity = datastore.Entity(key=key)
    print("entity: ", entity)
#     if req['queryResult']['action'] == "info-collect":
#         print("Query Text: ", req['queryResult']['queryText'])
#         print("Query Text 2: ", req.get('queryResult').get('fulfillmentText'))
#         entity["info-text"] = req['queryResult']['queryText']
#         print('Updated Entity value')
#     if req['queryResult']['action'] == 'pain-info':
#         print("Entity Value: ", entity["info-text"])
    
        
    params = req.get('queryResult').get('parameters')
    fulfillmentText = req.get('queryResult').get('fulfillmentText')
    if req['queryResult']['action'] == 'personal-info' and fulfillmentText.startswith("Yay!"):
        print('Request:')
        print(json.dumps(req, indent=4))
        data = {'name':str(params.get('person').get('name')), 
        'gender':str(params.get('gender')), 
        'age': int(params.get('age').get('amount')), 
        'phonenumber': str(params.get('phone-number')),
        'painintensity':4,
        'patientdesc': 'I am having rashes'} 
#         (url="http://96ab4d2e.ngrok.io/process-request?name=xyz&gender=male&age=65&patient_desc=I am having rashes&pain_intensity=4")
        API_ENDPOINT = "http://96ab4d2e.ngrok.io"+"/process-request" + "?" + "name="+data["name"] + "&gender="+data["gender"] + "&age=" + str(data["age"]) + "&phonenumber=" +  str(data["phonenumber"])+ "&painintensity=" +  str(data["painintensity"])+ "&patientdesc=" +  str(data["patientdesc"])
        
        print("data: ", data)
        requests.post(url = API_ENDPOINT)
        
    print(params)
    rsp = buildReply(fulfillmentText)
    rsp = json.dumps(rsp, indent=4)
#     print(rsp)
    r = make_response(rsp)
    r.headers['Content-Type'] = 'application/json'
    return r

def buildReply(info):
    return {
        'fulfillmentText': info,
    }

if __name__ == '__main__':
    app.run(host='0.0.0.0')
```
